```python
def Keff(kd,conversionrate,bradius,Keq=1e2):
    KD = kd*1e7 / 1e3# now is unit is M-1 S-1, need to convert to standunit mM-1 S-1 ; since mM = mol/m^3
    a = 1/(conversionrate[0]*1e-9) # since the lifetime is in ns
    #print a
    
    b = 1/(conversionrate[1]*1e-9)
    #print b
    R = (sum(bradius)/float(len(bradius)))*1e-10 # convert A to m
    #print KD
    #print R
    Diffusionconstant = KD/(4*3.1415*R*6.022e23*4e-4)
    #print Diffusionconstant
    td = R**2/Diffusionconstant
    #print td
    zab = 1 + ((a+b)*td)**0.5
    keff = KD* Keq * b*zab /( a*(Keq + KD*zab) + b*zab*(KD + Keq))
    print keff/KD
    return KD*1e3, keff*1e3 # now convert back to M-1 S-1
```
