```python
system = loos.createSystem(pdbfile)
#traj=loos.createTrajectory(dcdfile,system)

# readin the contact list
listfile="/net/share/bsu233/CaM/Go_model/ori/contactfile"
data=np.loadtxt(listfile)

scalafactor=1.5

Nlist=[]
Clist=[]
for i in np.arange(len(data[:,0])):
    if data[i,0] <= 75 and data[i,1] >= 146 :
        Nlist.append((data[i,0],data[i,1],data[i,3]))
    if data[i,0] >= 76 and data[i,0] <= 145 and data[i,1] >= 146 :
        Clist.append((data[i,0],data[i,1],data[i,3]))
    
#print Nlist,Clist       
 
C_termi = loos.selectAtoms(system,"resid >= 76 && resid <= 145")
N_termi = loos.selectAtoms(system,"resid >= 1 && resid <= 75")
pCaN = loos.selectAtoms(system,"resid >= 146")
#print pCaN
#for i,j in enumerate(Clist):
    #atom1 = loos.selectAtoms(system,"resid"+str(j[0]))
 #   print "resid = "+str(j[0])[0:-2]
```

```python
def CalculateContact(listinfo):
    traj=loos.createTrajectory(dcdfile,system)
    fraction=[]
    frame=0
    total=len(listinfo)*1.0
    while traj.readFrame():
        frame = frame + 1
        number = 0
        traj.updateGroupCoords(system)
        for i, j in enumerate(listinfo):
                str1 = "resid == "+str(j[0])[0:-2]
                str2 = "resid == "+str(j[1])[0:-2]
                atom1 = loos.selectAtoms(system,str1)
                atom2 = loos.selectAtoms(system,str2)
                centroid1 =  atom1.centroid() 
                centroid2 =  atom2.centroid()
                diff = centroid2 - centroid1
                distance = diff.length()
                #print distance,scalafactor*j[2]
                if distance <= scalafactor*j[2]:
                    number = number +1
       # print number
        
        fraction.append((frame,number/total))
                        
    return fraction

cc=CalculateContact(Clist) # list to store the contact information
nn = CalculateContact(Nlist) # list to store the contact information
    
    
```

```python
#print nn
xx,yy = zip(*cc)

#only draw the 400ns trajectory 
endframe = int(len(np.asanyarray(xx))*0.3)

xx1,yy1 = zip(*nn)# * means iterate throuh and use every tuples in cc as arguments
#print xx
plt.plot(np.asarray(xx)[0:endframe]*0.05,yy[0:endframe],"k.",label="Qc",markersize=1.5)
plt.plot(np.asarray(xx1)[0:endframe]*0.05,yy1[0:endframe],"r.",label="Qn",markersize=1.5)
plt.legend(fontsize=10,loc="lower right")
plt.xlabel("Simulation Time (ns)",fontsize=10,weight="bold")
plt.ylabel("Fraction of Native Contact",fontsize=10,weight="bold")
plt.xticks(fontsize=8,weight="bold")
plt.yticks([0.0,0.25,0.5,0.75,1.0],weight="bold",fontsize=8)
#plt.plot(xx1,yy1,"r.",lable="dd")
```

```python

def make_cmap(colors, position=None, bit=False):
    '''
    make_cmap takes a list of tuples which contain RGB values. The RGB
    values may either be in 8-bit [0 to 255] (in which bit must be set to
    True when called) or arithmetic [0 to 1] (default). make_cmap returns
    a cmap with equally spaced colors.
    Arrange your tuples so that the first color is the lowest value for the
    colorbar and the last is the highest.
    position contains values from 0 to 1 to dictate the location of each color.
    '''
    import matplotlib as mpl
    import numpy as np
    bit_rgb = np.linspace(0,1,256)
    if position == None:
        position = np.linspace(0,1,len(colors))
    else:
        if len(position) != len(colors):
            sys.exit("position length must be the same as colors")
        elif position[0] != 0 or position[-1] != 1:
            sys.exit("position must start with 0 and end with 1")
    if bit:
        for i in range(len(colors)):
            colors[i] = (bit_rgb[colors[i][0]],
                         bit_rgb[colors[i][1]],
                         bit_rgb[colors[i][2]])
    cdict = {'red':[], 'green':[], 'blue':[]}
    for pos, color in zip(position, colors):
        cdict['red'].append((pos, color[0], color[0]))
        cdict['green'].append((pos, color[1], color[1]))
        cdict['blue'].append((pos, color[2], color[2]))

    cmap = mpl.colors.LinearSegmentedColormap('my_colormap',cdict,256)
    return cmap

#colors = [(0,0,255), (255,255,51), (255,0,0), (255,255,255)] # This example uses the 8-bit RGB
### Call the function make_cmap which returns your colormap
#my_cmap = make_cmap(colors, bit=True)
### Create an array or list of positions from 0 to 1.
#position = [0, 0.5,0.7 ,1]
#tmap=make_cmap(colors, position=position)
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
vmax = 7.0
cmap = LinearSegmentedColormap.from_list('mycmap', [(0 / vmax, 'darkblue'),
                                                    (2 / vmax, 'aqua'),
                                                    (4 / vmax, 'yellow'),
                                                    (6.999 / vmax, 'r'),
                                                    (7 / vmax, 'white')]
                                        )

plt.imshow(newdata,vmax=7,cmap=cmap,interpolation=None)

lab=np.arange(0,21,5)
ax = plt.gca()
ax.invert_yaxis()


cct=plt.colorbar()
plt.xticks(lab,lab*0.05,fontsize=8)
plt.yticks(lab,lab*0.05,fontsize=8)
plt.xlabel("$Q_{N}$",fontsize=12,weight="bold")
plt.ylabel("$Q_{C}$",fontsize=12,weight="bold")
#plt.clabel("$Q_{C}$",fontsize=12,weight="bold")
plt.xticks(fontsize=8,weight="bold")
plt.yticks(fontsize=8,weight="bold")
cct.ax.tick_params(labelsize=8)
cct.ax.set_ylabel('$k_b T$', fontsize=12,weight="bold")


# create an axes on the right side of ax. The width of cax will be 5%
# of ax and the padding between cax and ax will be fixed at 0.05 inch.
#divider = make_axes_locatable(ax)
#cax = divider.append_axes("right", size="5%", pad=0.05)
```

```python
def BD_contact(system,traj,listinfo):
    fraction=[]
    frame=0
    total=len(listinfo)*1.0
    while traj.readFrame():
        frame = frame + 1
        number = 0
        traj.updateGroupCoords(system)
        for i, j in enumerate(listinfo):
                str1 = 'chainid == "A" && name == "CA" && resid == '+str(j[0])[0:-2]
                str2 = 'chainid == "B" && name == "CA" && resid == '+str(j[1])[0:-2]
                #print str2
                atom1 = loos.selectAtoms(system,str1)
                atom2 = loos.selectAtoms(system,str2)
                centroid1 =  atom1.centroid() 
                centroid2 =  atom2.centroid()
                diff = centroid2 - centroid1
                distance = diff.length()
                #print distance
                #print distance,scalafactor*j[2]
                if distance <= 2*j[2]: #Note, for BD sims, the scalafoctor is set as 2 instead of 1.5
                    number = number +1
       # print number
        
        fraction.append((frame,number/total))
                        
    return fraction

BD_N_fraction = BD_contact(nsystem,BD_N_traj,BD_Nlist)
BD_C_fraction = BD_contact(csystem,BD_C_traj,BD_Clist)

closed_BD_N_fraction = BD_contact(closed_nsystem,closed_BD_N_traj,BD_Nlist)
closed_BD_C_fraction = BD_contact(closed_csystem,closed_BD_C_traj,BD_Clist)
```
