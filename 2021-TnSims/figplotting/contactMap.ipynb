```python
import numpy as np
import matplotlib.pylab as plt
```

```python
datadir = "/home/binsun/TnC/"
cases = ['1J1E','1J1E_C84Y','6KN7','6KN7_C84Y','6KN8','6KN8_C84Y'] 
rawData = {i:np.loadtxt(datadir+i+"/alltraj/contactmap/"+"nonnative.cmap",comments="#",usecols=2) for i in cases}
```

```python
contacts = {}
for i in cases:
    contacts[i] = rawData[i].reshape((160,160))
```

```python
contactData = {}
for i in cases:
    contactData[i]= np.moveaxis(contacts[i],source=0,destination=1)
```

```python
Case='1J1E'
sns.set(font_scale = 0.5)
mask = np.zeros_like(contactData[Case])
#mask[np.triu_indices_from(mask)] = True
yticks=np.linspace(0,160,21,dtype=np.int) #yticks=np.asarray([0,49,144,195,220])
xticks=np.linspace(0,160,21)
xticklabels = xticks + 1
#yticklabels = np.from(yticks + 1).tostring()


#print (xticklabels)
#yticklabels = yticks+441
with sns.axes_style("white"):
    ax = plt.figure(figsize=(4, 2),dpi=300)
    ax = sns.heatmap(contactData[Case], center=10, square=True,vmax=20,\
                    yticklabels=xticklabels,xticklabels=xticklabels)
    #ax.set_xlim([1,145])
    #ax.set_ylim([196,221])
    ax.invert_yaxis()
    ax.set_yticks(yticks)
    ax.set_xticks(xticks)
    ax.set_ylabel('TnC',fontsize=8,labelpad=0)
    ax.set_xlabel("TnC",fontsize=8,labelpad=0)
    
    ax.tick_params(axis='both', which='major', pad=-5)
    #ax.xaxis.set_minor_locator(MaxNLocator(integer=True))
    ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))
    ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
    
    #ax.colorbar(False)
    #ax.label_params(axis='both', which='major', pad=-5)
    
```
