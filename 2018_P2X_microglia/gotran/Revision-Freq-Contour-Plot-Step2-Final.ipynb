```python
from scipy.interpolate import griddata
grid_x, grid_y = np.mgrid[0.5:5:100j, 0:2:100j]

I = np.array([0.5, 1, 2, 4, 5])
J = np.array([0, 0.01, 0.1, 1, 2])

n = 0

for i in I:
    for j in J:
        pairnew = [i,j]
        if n == 0:
            points = pairnew
        else:
            points = np.vstack((points,pairnew))
        n = n + 1

Z = np.array([ 0.10003805,  0.10015645,  0.10089604,  0.10161495,  0.10245972,
  0.10003805,  0.10021328,  0.10163987,  0.10262356,  0.10425218,
  0.10003805,  0.10032443,  0.10452228,  0.10541397,  0.10932511,
  0.10003805,  0.10065164,  0.11404071,  0.11457998,  0.12539467,
  0.10003805,  0.10081747,  0.11775971,  0.11837237,  0.13183332])
plt.figure(figsize=(7,8),dpi=200)
grid_z0 = griddata(points, Z, (grid_x, grid_y),method='linear')
im=plt.imshow(grid_z0.T, extent=(0.5,5,0,2), origin='lower',interpolation='bilinear',
                cmap=cm.gray, aspect='auto')

CS=plt.contour(grid_z0.T, 0.11,
            origin='lower',
            colors='gold',
            linewidths=2,
           extent=(0.5,5,0,2))

CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8)
CBI.set_label('$[Ca^{2+}]_{i}$ conc. (uM)', fontsize=12)
plt.clabel(CS, inline=1, fontsize=12)
plt.xlabel("Frequency (Hz)\n (0.1 sec of stimulation)",fontsize=15)
plt.tight_layout()
plt.ylabel("ATP conc. (mM)",fontsize=15)
```
