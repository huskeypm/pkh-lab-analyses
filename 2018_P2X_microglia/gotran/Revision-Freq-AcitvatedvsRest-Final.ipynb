```python
ca1a = ao.GetData(data1a,"Cai") # 2 Activated 0.01
ca1b = ao.GetData(data1b,"Cai") # 4 
ca1c = ao.GetData(data1c,"Cai") # 5

tnf1a = ao.GetData(data1a,"TNFa")
tnf1b = ao.GetData(data1b,"TNFa")
tnf1c = ao.GetData(data1c,"TNFa")

ca2a = ao.GetData(data2a,"Cai")
ca2b = ao.GetData(data2b,"Cai")
ca2c = ao.GetData(data2c,"Cai")

tnf2a = ao.GetData(data2a,"TNFa")
tnf2b = ao.GetData(data2b,"TNFa")
tnf2c = ao.GetData(data2c,"TNFa")

ca3a = ao.GetData(data3a,"Cai")
ca3b = ao.GetData(data3b,"Cai")
ca3c = ao.GetData(data3c,"Cai")

tnf3a = ao.GetData(data3a,"TNFa")
tnf3b = ao.GetData(data3b,"TNFa")
tnf3c = ao.GetData(data3c,"TNFa")

ca4a = ao.GetData(data4a,"Cai")
ca4b = ao.GetData(data4b,"Cai")
ca4c = ao.GetData(data4c,"Cai")

tnf4a = ao.GetData(data4a,"TNFa")
tnf4b = ao.GetData(data4b,"TNFa")
tnf4c = ao.GetData(data4c,"TNFa")

ca1aA10 = np.mean(ca1a.valsIdx[10000:20000]) # 2 Activated 0.01
ca1bA10 = np.mean(ca1b.valsIdx[10000:20000]) # 4 Activated 0.01
ca1cA10 = np.mean(ca1c.valsIdx[10000:20000]) # 5 Activated 0.01

ca2aA100 = np.mean(ca2a.valsIdx[10000:20000]) # 2 Activated 0.1
ca2bA100 = np.mean(ca2b.valsIdx[10000:20000]) # 4 Activated 0.1
ca2cA100 = np.mean(ca2c.valsIdx[10000:20000]) # 5 Activated 0.1

ca3aR10 = np.mean(ca3a.valsIdx[10000:20000]) # 2 Rest 0.01
ca3bR10 = np.mean(ca3b.valsIdx[10000:20000]) # 4 Rest 0.01
ca3cR10 = np.mean(ca3c.valsIdx[10000:20000]) # 5 Rest 0.01

ca4aR100 = np.mean(ca4a.valsIdx[10000:20000]) # 2 Rest 0.1
ca4bR100 = np.mean(ca4b.valsIdx[10000:20000]) # 4 Rest 0.1
ca4cR100 = np.mean(ca4c.valsIdx[10000:20000]) # 5 Rest 0.1


tnf1aA10 = (tnf1a.valsIdx[20000]-tnf1a.valsIdx[10000])/(0.167)
tnf1bA10 = (tnf1b.valsIdx[20000]-tnf1b.valsIdx[10000])/(0.167)
tnf1cA10 = (tnf1c.valsIdx[20000]-tnf1c.valsIdx[10000])/(0.167)

tnf2aA100 = (tnf2a.valsIdx[20000]-tnf2a.valsIdx[10000])/(0.167)
tnf2bA100 = (tnf2b.valsIdx[20000]-tnf2b.valsIdx[10000])/(0.167)
tnf2cA100 = (tnf2c.valsIdx[20000]-tnf2c.valsIdx[10000])/(0.167)

tnf3aR10 = (tnf3a.valsIdx[20000]-tnf3a.valsIdx[10000])/(0.167)
tnf3bR10 = (tnf3b.valsIdx[20000]-tnf3b.valsIdx[10000])/(0.167)
tnf3cR10 = (tnf3c.valsIdx[20000]-tnf3c.valsIdx[10000])/(0.167)

tnf4aR100 = (tnf4a.valsIdx[20000]-tnf4a.valsIdx[10000])/(0.167)
tnf4bR100 = (tnf4b.valsIdx[20000]-tnf4b.valsIdx[10000])/(0.167)
tnf4cR100 = (tnf4c.valsIdx[20000]-tnf4c.valsIdx[10000])/(0.167)

tnfatotal = np.array([tnf1aA10,tnf1bA10,tnf1cA10,
                      tnf2aA100,tnf2bA100,tnf2cA100,
                      tnf3aR10,tnf3bR10,tnf3cR10,
                      tnf4aR100,tnf4bR100,tnf4cR100]) 
Tmax = max(tnfatotal)

ind1 = np.array([0.9,1.9,2.9])
ind2 = np.array([1.1,2.1,3.1])
ind = np.array([1,2,3])

caA10 = np.array([ca1aA10,ca1bA10,ca1cA10])
caR10 = np.array([ca3aR10,ca3bR10,ca3cR10])

caA100 = np.array([ca2aA100,ca2bA100,ca2cA100])
caR100 = np.array([ca4aR100,ca4bR100,ca4cR100])

tnfA10 = np.array([tnf1aA10,tnf1bA10,tnf1cA10])/Tmax
tnfR10 = np.array([tnf3aR10,tnf3bR10,tnf3cR10])/Tmax

tnfA100 = np.array([tnf2aA100,tnf2bA100,tnf2cA100])/Tmax
tnfR100 = np.array([tnf4aR100,tnf4bR100,tnf4cR100])/Tmax

plt.figure(figsize=(5, 5), dpi=200)
ax = plt.subplot(1,1,1)
ax.tick_params(labelsize=12)
ax.set_title('10 uM ATP',fontsize=12)
R1,R2,R3 = ax.bar(ind2, caR10, width=0.2, align='center', label='rest')
R1.set_facecolor('b')
R2.set_facecolor('b')
R3.set_facecolor('b')
A1,A2,A3 = ax.bar(ind1, caA10, width=0.2, align='center', label='act')
A1.set_facecolor('r')
A2.set_facecolor('r')
A3.set_facecolor('r')
ax.set_xticks(ind)
ax.set_ylim([0.1, 0.22])
ax.set_ylabel('Ave. Ca Conc. (uM)',fontsize=12)
ax.set_xticklabels(['2','4','5'])
ax.set_xlabel('Frequency (Hz)',fontsize=12)
ax.legend(loc=0,fontsize=10)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
plt.tight_layout()

plt.figure(figsize=(5, 5), dpi=200)
ax = plt.subplot(1,1,1)
ax.tick_params(labelsize=12)
ax.set_title('10 uM ATP',fontsize=12)
R1,R2,R3 = ax.bar(ind2, tnfR10, width=0.2, align='center', label='rest')
R1.set_facecolor('b')
R2.set_facecolor('b')
R3.set_facecolor('b')
A1,A2,A3 = ax.bar(ind1, tnfA10, width=0.2, align='center', label='act')
A1.set_facecolor('r')
A2.set_facecolor('r')
A3.set_facecolor('r')
ax.set_xticks(ind)
ax.set_ylim([0, 1])
ax.set_ylabel('TNFa syn. rate (frac. of max)',fontsize=12)
ax.set_xticklabels(['2','4','5'])
ax.set_xlabel('Frequency (Hz)',fontsize=12)
ax.legend(loc=0,fontsize=10)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
plt.tight_layout()

plt.figure(figsize=(5, 5), dpi=200)
ax = plt.subplot(1,1,1)
ax.tick_params(labelsize=12)
ax.set_title('100 uM ATP',fontsize=12)
R1,R2,R3 = ax.bar(ind2, caR100, width=0.2, align='center', label='rest')
R1.set_facecolor('b')
R2.set_facecolor('b')
R3.set_facecolor('b')
A1,A2,A3 = ax.bar(ind1, caA100, width=0.2, align='center', label='act')
A1.set_facecolor('r')
A2.set_facecolor('r')
A3.set_facecolor('r')
ax.set_xticks(ind)
ax.set_ylim([0.1, 0.22])
ax.set_ylabel('Ave. Ca Conc. (uM)',fontsize=12)
ax.set_xticklabels(['2','4','5'])
ax.set_xlabel('Frequency (Hz)',fontsize=12)
ax.legend(loc=0,fontsize=10)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
plt.tight_layout()

plt.figure(figsize=(5, 5), dpi=200)
ax = plt.subplot(1,1,1)
ax.tick_params(labelsize=12)
ax.set_title('100 uM ATP',fontsize=12)
R1,R2,R3 = ax.bar(ind2, tnfR100, width=0.2, align='center', label='rest')
R1.set_facecolor('b')
R2.set_facecolor('b')
R3.set_facecolor('b')
A1,A2,A3 = ax.bar(ind1, tnfA100, width=0.2, align='center', label='act')
A1.set_facecolor('r')
A2.set_facecolor('r')
A3.set_facecolor('r')
ax.set_xticks(ind)
ax.set_ylim([0, 1])
ax.set_ylabel('TNFa syn. rate (frac. of max)',fontsize=12)
ax.set_xticklabels(['2','4','5'])
ax.set_xlabel('Frequency (Hz)',fontsize=12)
ax.legend(loc=0,fontsize=10)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
plt.tight_layout()
```
