```python
# Function for integreating ODE and displaying results
import scipy.integrate
from scipy.integrate import odeint
import matplotlib
import matplotlib.pylab as plt
matplotlib.rcParams['figure.figsize'] = (6,3.75)
matplotlib.rcParams['savefig.dpi'] = 120
import numpy as np
```

```python
def f(y,t,a):
  return dot(a,y)
#    return [a[0,0]*y[0] + a[0,1]*y[1],\
#            a[1,0]*y[0] + a[1,1]*y[1]]


t  = scipy.linspace(0.,10.,100)
```

```python
# Initial conc. 
y0 = [2,3]

# Jacobian

kf3=0.8
kb3=0.2
kf1=1.9
kb1=0.2
kf2=0.9
kb2=0.8
kf4=1.6
kb4=0.9

K1=kf1/kb1
K3=kf3/kb3

af1=kf4*K3/(1+K3)
ab1=kb4/(1+K1)
af2=kf2*K1/(1+K1)
ab2=kb2/(1+K3)

J1 = np.array([
  [-(kf3+kb2),kb3,0,kf2],
  [kf3,-(kb3+kf4),kb4,0],
  [0,kf4,-(kb4+kf1),kb1],
  [kb2,0,kf1,-(kf2+kb1)]
 ]) 

J2 = np.array([
  [-(af1+ab2),ab1+af2],
  [af1+ab2,-(ab1+af2)],
 ]) 


y1 = 1
y2 = 0
y3 = 0
y4 = 0

ly1=1
ly2=0

ys = np.array([y1,y2,y3,y4],dtype=np.float)
daSum = np.sum(ys)
y = scipy.integrate.odeint(f,ys,t,args=(J1,))

lys = np.array([ly1,ly2],dtype=np.float)
ldaSum = np.sum(lys)
lumy = scipy.integrate.odeint(f,lys,t,args=(J2,))



#print y[:,1]
#print y[:,0]
#print y[:,0]+y[:,1]-lumy[:,0]
#print y[:,2]+y[:,3]-lumy[:,1]
#print lumy[:,0]
plot(t,y[:,0],"r.",label="y$_1$")
plot(t,y[:,1],"r-.",label="y$_2$")
plot(t,y[:,2],"b.",label="y$_3$")
plot(t,y[:,3],"b-.",label="y$_4$")
plot(t,lumy[:,0],"r-",label="E$_1$")
plot(t,lumy[:,1],"b-",label="E$_2$")
ylabel("$[y_i]$")
xlabel("t")
legend(loc=0)
```

```python
# two state model ( only Ca2+ ) turn over rate 
# in Tran's paper, for two-state model, the expression of turin over is:
#  V= (af1*af2-ab1*ab2)/(ab1+ab2+af1+af2)

# experimental data from Cantilina's paper (figure 3A, nomalized to ensure maxmium is 5/S-1)
c0=[7.76046e-08,9.94839e-08,1.92927e-07,2.97960e-07,4.79622e-07,7.72041e-07,9.30121e-07,1.88000e-06,4.67372e-06,9.64428e-06]
r0=[0.272727,0.285714,0.636364,0.870130,2.10390,2.74026,4,4.72727,4.84416,4.94805]
                   
def cRate(Ca,kf5,kb5,kf6,kb6):
    Kd=0.00224 # dissociation constance of Calcium in the SR side, taken from Tran's paper
    #Kdk=0.0253
    MgATP=0.0003
    Pi=0.000001
    #Ca=0.000005
    MgADP=0.000001

    Cai=0.000001
    kf1=100000000.0
    kb1=400.0
    kf2=30.0
    kb2=40.0
    kf3=400000000.0
    kb3=16.0
    kf4=20000000.0
    kb4=40.0
   
    
    K1=kf1/kb1
    K3=kf3/kb3
    K2=kf2/kb2
    #K3=kf3/kb3
    K4=kf4/kb4
     
    af1=kf5*K1*K2*K3*K4*Ca*Ca*MgATP/(1+K1*Ca+K1*K2*Ca+K1*K2*K3*Ca*Ca+K1*K2*K3*K4*Ca*Ca*MgATP)
    ab1=kb5*MgADP/(1+Kd/Cai+Kd*Kd/(Cai*Cai))
    af2=(kf6*Kd*Kd/(Cai*Cai))/(1+(Kd/Cai)+Kd*Kd/(Cai*Cai))
    ab2=kb6*Pi/(1+K1*Ca+K1*K2*Ca+K1*K2*K3*Ca*Ca+K1*K2*K3*K4*Ca*Ca*MgATP)
    
    return (af1*af2-ab1*ab2)/(ab1+ab2+af1+af2)
    
vcRate = np.vectorize(cRate)
# Both Mg and K are inhibtors of SERCA
def MgKin_Rate(Ca,Mg,Kdmg,K,Kdk):

    Kd=0.00224
    #Kdk=1.2
    #Kdmg=0.1
    MgATP=0.0003
    Pi=0.000001
    #Ca=0.000005
    MgADP=0.000001

    Cai=0.000001
    kf1=100000000.0
    kb1=400.0
    kf2=30.0
    kb2=40.0
    kf3=400000000.0
    kb3=16.0
    kf4=20000000.0
    kb4=40.0
    kf5=15
    kb5=75000
    kf6=7.5
    kb6=5000000000
    K1=kf1/kb1
    K3=kf3/kb3
    K2=kf2/kb2
    K4=kf4/kb4
    
    ad1=Kdmg/Mg
    ad2=Kdk/K
    
    af1=kf5*ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP/(1+ad1+ad1*ad2+ad1*ad2*K1*Ca+ad1*ad2*K1*K2*Ca+ad1*ad2*K1*K2*K3*Ca*Ca+ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP)
    ab1=kb5*MgADP/(1+Kd/Cai+Kd*Kd/(Cai*Cai))
    af2=(kf6*Kd*Kd/(Cai*Cai))/(1+(Kd/Cai)+Kd*Kd/(Cai*Cai))
    ab2=kb6*Pi/(1+ad1+ad1*ad2+ad1*ad2*K1*Ca+ad1*ad2*K1*K2*Ca+ad1*ad2*K1*K2*K3*Ca*Ca+ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP)
    
    return (af1*af2-ab1*ab2)/(ab1+ab2+af1+af2)
    
vMgKin_Rate = np.vectorize(MgKin_Rate)

def MgIKT_Rate(Ca,Mg,Kdmg,K,Kdk):
    Kd=0.00224
   
    #Kdk=0.1
    #Kdmg=0.003
    
    MgATP=0.0003
    Pi=0.000001
    #Ca=0.000005
    MgADP=0.000001

    Cai=0.000001
    
    
    Cai=0.000001
    kf1=100000000.0
    kb1=400.0
    kf2=30.0
    kb2=40.0
    kf3=400000000.0
    kb3=16.0
    kf4=20000000.0
    kb4=40.0
    kf5=15
    kb5=7.5
    kf6=7.5
    kb6=7000000000
    K1=kf1/kb1
    K3=kf3/kb3
    K2=kf2/kb2
    K4=kf4/kb4
    ad1=Kdmg/Mg
    ad2=Kdk/K
    
    af1=kf5*ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP/(1+ad2+ad1*ad2+ad1*ad2*K1*Ca+ad1*ad2*K1*K2*Ca+ad1*ad2*K1*K2*K3*Ca*Ca+ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP)
    ab1=kb5*MgADP/(1+Kd/Cai+Kd*Kd/(Cai*Cai))
    af2=(kf6*Kd*Kd/(Cai*Cai))/(1+(Kd/Cai)+Kd*Kd/(Cai*Cai))
    ab2=kb6*Pi/(1+ad2+ad1*ad2+ad1*ad2*K1*Ca+ad1*ad2*K1*K2*Ca+ad1*ad2*K1*K2*K3*Ca*Ca+ad1*ad2*K1*K2*K3*K4*Ca*Ca*MgATP)
    
    return (af1*af2-ab1*ab2)/(ab1+ab2+af1+af2)
    
vMgIKT_Rate = np.vectorize(MgIKT_Rate)

cs = 10**np.linspace(-8,-4,100)
#print vcRate(cs,0.0000002)
#plt.plot(cs,vcRate(cs,15,75000,250,270000),"k-",label="Only Ca2+")
plt.plot(cs,vcRate(cs,15,75000,7.5,5000000000),"b-",label="Only Ca2+") # Now we know the value of K5 and K6
plt.plot(c0,r0,"k.",label="Experimental data")
plt.plot(cs,vMgKin_Rate(cs,0.002,0.003,0.15,0.1),"m.",label="Only Ca2+")
#plt.plot(cs,vMgIKT_Rate(cs,0.002,0.15),"r.",label="Only Ca2+")

xscale('log')
plt.legend(loc=0)
plt.xlabel("[Ca]")
plt.ylabel("Pump Rate(1/S)")
```

```python
# Turn over rate for Mg and K both are inhibitors
# Mg and K are both intermediates
def MgKTT_Rate(Ca,Mg,Kdmg,K,Kdk):
    Kd=0.00224
   
    #Kdk=0.1
    #Kdmg=0.003
    
    MgATP=0.0003
    Pi=0.000001
    #Ca=0.000005
    MgADP=0.000001

  
    
    Cai=0.000001
    kf1=100000000.0
    kb1=400.0
    kf2=30.0
    kb2=40.0
    kf3=400000000.0
    kb3=16.0
    kf4=20000000.0
    kb4=40.0
    kf5=15
    kb5=7.5
    kf6=7.5
    kb6=7000000000
    K1=kf1/kb1
    K3=kf3/kb3
    K2=kf2/kb2
    K4=kf4/kb4
  
   
    kf0=1/(1/Kdmg+1/Mg+K/(Mg*Kdk))
    
    
    ad1=kf0/Mg
    
    
    af1=kf5*ad1*K1*K2*K3*K4*Ca*Ca*MgATP/(1+ad1+ad1*K1*Ca+ad1*K1*K2*Ca+ad1*K1*K2*K3*Ca*Ca+ad1*K1*K2*K3*K4*Ca*Ca*MgATP)
    ab1=kb5*MgADP/(1+Kd/Cai+Kd*Kd/(Cai*Cai))
    af2=(kf6*Kd*Kd/(Cai*Cai))/(1+(Kd/Cai)+Kd*Kd/(Cai*Cai))
    ab2=kb6*Pi/(1+ad1+ad1*K1*Ca+ad1*K1*K2*Ca+ad1*K1*K2*K3*Ca*Ca+ad1*K1*K2*K3*K4*Ca*Ca*MgATP)
    
    return (af1*af2-ab1*ab2)/(ab1+ab2+af1+af2)
    
vMgKTT_Rate = np.vectorize(MgKTT_Rate)

plt.plot(cs,vcRate(cs,15,7.5,7.5,7000000000),"c-",label="Only Ca2+") # Now we know the value of K5 and K6
plt.plot(c0,r0,"k.",label="Experimental data")
plt.plot(cs,vMgKin_Rate(cs,0.002,0.0015,0.15,0.3),"r-",label="MgIKI")
plt.plot(cs,vMgIKT_Rate(cs,0.002,0.0015,0.15,0.3),"b-",label="MgIKT")
#plt.plot(cs,vMgKTT_Rate(cs,0.002,0.005,0.15,0.2),"b-",label="MgTKT")
plt.plot(cs,vMgKTT_Rate(cs,0.002,0.0015,0.15,0.3),"g-",label="MgTKT")
xscale('log')
plt.legend(loc=0)
plt.xlabel("[Ca]")
plt.ylabel("Pump Rate(1/S)")
```
