## Calculations

Here we compute the ideal scaling, assuming your first data point (lowest number of processors) is perfectly efficient

```python
idealSingleProc = myResults[0]/numProcs[0]
print idealSingleProc, myMetric

idealResults = idealSingleProc*numProcs
pctEfficiencyTol = 0.8
tolResults = idealResults * pctEfficiencyTol
```

Find maximum number of processors for which efficiency is greater than cutoff

```python
argEff = np.argwhere(myResults> (pctEfficiencyTol*idealResults))
maxPt = argEff[-1]
print maxPt
```

```python
lg2 = np.log(numProcs)/np.log(2)
#lg2=numProcs
myResultslg2 = np.log(myResults)/np.log(2)
#myResultslg2 = myResults

ideallg2 = np.log(idealResults)/np.log(2)
#ideallg2 = idealResults

tollg2 = np.log(tolResults)/np.log(2)
#tollg2 =tolResults

maxLg = myResultslg2[maxPt]
#maxLg = myResultslg2[maxPt]

plt.plot(lg2,myResultslg2,'b.',label="actual")
plt.plot(lg2,ideallg2,'b-',label="ideal")
plt.plot(lg2,tollg2,'r--',label="%3.1f*ideal"%pctEfficiencyTol)
plt.plot(lg2[maxPt],myResultslg2[maxPt],'r.',markersize=20,label="Best value\n(%d procs)\n(%4.2f %s)"%(numProcs[maxPt],myResults[maxPt], myMetric))

#yscale('log')
plt.xlabel("processors (lg2)")
plt.ylabel("%s (lg2)"%myMetric) 
plt.legend(loc=2)
plt.title("%s/%s"%(caseName,resource))

plt.gcf().savefig(figureName,dpi=300)
```
