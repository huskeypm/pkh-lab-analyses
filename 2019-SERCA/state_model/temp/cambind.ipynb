$$\phi_f = \frac{1}{1 + c^3 K_{12}(1+c K_{24}(1+p K_{45}(1+c K_{35}^{-1})))}$$
$$\phi_f = \frac{1}{1 + c^3 K_{12}(1+c K_{24}))}$$


$$\phi_{bound} \equiv 1-\phi_f$$

```python
K12=1.e18; K45=1e1; K35=1e0;
```

```python
def func(c,p,K45=1e7):
    denom = 1 + c**3*K12*1+c*K24*(1+p*K45*(1+c*K35**(-1)))
    return 1 - 1/denom 

vfunc = np.vectorize(func)
cs = 10**np.linspace(-9,-5,100)
```

```python
p=0.; K45=1e1;
plt.plot(cs,vfunc(cs,p,K45),label="p = %3.1f [M], K$_{45}$ = %3.1f[1/M]"%(p,K45))
p=0.1; K45=1e1;
plt.plot(cs,vfunc(cs,p,K45),label="p = %3.1f [M], K$_{45}$ = %3.1f[1/M]"%(p,K45))
p=0.1; K45=2e1;
plt.plot(cs,vfunc(cs,p,K45),label="p = %3.1f [M], K$_{45}$ = %3.1f[1/M]"%(p,K45))
ylim([0,1])
xscale('log')
plt.legend(loc=0)
plt.xlabel("[Ca]")
plt.ylabel("[CaCaM]")
plt.title("Ca-bound CaM probability wrt\n CaM target peptide concentration and peptide affinity")
plt.gcf().savefig("/home/AD/pmke226/grants/151101_EF_NSF/figs/cambind.png")
```
