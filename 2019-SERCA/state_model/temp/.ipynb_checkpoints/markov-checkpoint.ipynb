#### Simplest system - monoexponential decay

$dy/dt = -a \cdot y$

Solution:
$y = c e^{-at}$

```python
def f(y,t,a):
    dydt = -a*y[0]
    return dydt

# points in time where the function gets evaluated, t[0] at y0
t  = scipy.linspace(0.,10.,201)
# init conc
y0 = [16.]

# rate
a=0.5  # 
y = scipy.integrate.odeint(f,y0,t,args=(a,))
```

Below are the data from the class demonstration

```python
# class data

yClass = [16.,8.,4.,2.,1.]
tClass = np.arange(5)  # t=0,1,2,3,4
```

Here we compare results. Why do the class results deviate from the analytical solution? 

```python
plt.plot(t,y,'k.',label="Numerical result")

analyticalResult = y[0]*np.exp(-a*t)
plt.plot(t,analyticalResult,label="Analytical result")

plt.plot(tClass,yClass,'r.',label="Class")

plt.ylabel("C")
plt.xlabel("t")
plt.legend()
plt.show()
```

#### Simple system - two components

\dot{y_1} &= -a_{11} y_1 \\

Using Jacobian:

First verifying that diagonal Jacobian with negative rate constants yields exponential decay

$\frac{dy[0]}{dt} = a\cdot y[0] +b\cdot y[1]$

$\frac{dy[1]}{dt} = c\cdot y[0]+d\cdot y[1]$

#### Multi-component system with transition probability matrix J

Using Jacobian:
 ( $a_{ij} = \frac{dy_i}{da_j}$)
 
 

```python
# turn over rate

# only Ca2+ included
def Rateonly_Ca(Ca,MgATP,MgADP,Pi,kf1,kb1,kf2,kb2,kf3,kb3,kf4,kb4,kf5,kb5,kf6,kb6):
    K1=kf1/kb1
    K2=kf2/kb2
    K3=kf3/kb3

    af1=(kf4*MgATP)/(1+(1/(K1*K2*K3*Ca*Ca))+1/(K2*K3*Ca)+1/(K3*Ca))
    ab1=kb4
    af2=kf5
    ab2=kb5*MgADP
    af3=kf6
    ab3=((kb6*Pi)/(1+K1*Ca+K1*K2*Ca+K1*K2*K3*Ca))
    
    nom = af1*af2*af3-ab1*ab2*ab3
    denom =(af2*af3+ab1*af3+ab1*ab2+af1*af3+ab2*af1+ab2*ab3+af1*af2+ab3*ab1+ab3*af2)
    
    return  nom/denom

vRateonly_Ca = np.vectorize(Rateonly_Ca)

# when Mg2+ act as an inhibitor
def Rate_Mginhibitor(Ca,Kdmg,Mg,MgATP,MgADP,Pi,kf1,kb1,kf2,kb2,kf3,kb3,kf4,kb4,kf5,kb5,kf6,kb6):
    K1=kf1/kb1
    K2=kf2/kb2
    K3=kf3/kb3

    af1=(kf4*MgATP)/(1+Mg/(Kdmg*K1*K2*K3*Ca*Ca)+1/(K1*K2*K3*Ca*Ca)+1/(K2*K3*Ca)+1/(K3*Ca))
    ab1=kb4
    af2=kf5
    ab2=kb5*MgADP
    af3=kf6
    ab3=(kb6*Pi)/(1+Kdmg/Mg+(Kdmg*K1*Ca)/Mg+(Kdmg*K1*K2*Ca)/Mg+(Kdmg*K1*K2*K3*Ca*Ca)/Mg)
    
    nom = af1*af2*af3-ab1*ab2*ab3
    denom =(af2*af3+ab1*af3+ab1*ab2+af1*af3+ab2*af1+ab2*ab3+af1*af2+ab3*ab1+ab3*af2)
    
    return  nom/denom

vRate_Mginhibitor = np.vectorize(Rate_Mginhibitor)

# when Mg2+ act as an inhibitor and K+ is an intermediate 
def Rate_MginKinter(Ca,Kdmg,Mg,Kdk,Ka,MgATP,MgADP,Pi,kf1,kb1,kf2,kb2,kf3,kb3,kf4,kb4,kf5,kb5,kf6,kb6):
    K1=kf1/kb1
    K2=kf2/kb2
    K3=kf3/kb3

    af1=(kf4*MgATP)/(1+(Mg*Ka)/(Kdmg*Kdk*K1*K2*K3*Ca*Ca)+Ka/(Kdk*K1*K2*K3*Ca*Ca)+1/(K1*K2*K3*Ca*Ca)+1/(K2*K3*Ca)+1/(K3*Ca))
    ab1=kb4
    af2=kf5
    ab2=kb5*MgADP
    af3=kf6
    ab3=(kb6*Pi)/(1+Kdmg/Mg+(Kdmg*Kdk)/(Mg*Ka)+(Kdmg*Kdk*K1*Ca)/(Mg*Ka)+(Kdmg*Kdk*K1*K2*Ca)/(Mg*Ka)+(Kdmg*Kdk*K1*K2*K3*Ca*Ca)/(Mg*Ka))
    
    nom = af1*af2*af3-ab1*ab2*ab3
    denom =(af2*af3+ab1*af3+ab1*ab2+af1*af3+ab2*af1+ab2*ab3+af1*af2+ab3*ab1+ab3*af2)
    
    return  nom/denom

vRate_MginKinter = np.vectorize(Rate_MginKinter)
print vRate_MginKinter(cs,0.0253,0.02,0.0253,0.15,MgATP,MgADP,Pi,kf1,kb1,kf2,kb2,kf3,kb3,kf4,kb4,kf5,kb5,kf6,kb6)
```
