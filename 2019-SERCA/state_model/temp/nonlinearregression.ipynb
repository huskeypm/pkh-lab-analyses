## Example of linear regression

```python
eps_data = 1.

# y = a exp(-t/tau)
def GenData(vars,x):
    amp = vars[0]
    decay = vars[1]

    return amp * np.exp(-x/decay)


## Generate data
vars = [1.2,500.]
print vars
x = np.linspace(0,1000,100)
data = GenData(vars,x) + 0.05*np.random.randn(np.shape(x)[0])

# log y = log a - t/tau
y = log(data)
regression = np.polyfit(x, y, 1)
aFit = np.exp(regression[1])
tauFit = -1/regression[0]
varsFit = [aFit,tauFit]
print varsFit


## Compare
dataFit = GenData(varsFit,x)

plt.figure()
plt.plot(x,data,label="data")
plt.plot(x,dataFit,label="Fit")
plt.legend()
```

## Example with nonlinear regression

```python
eps_data = 1.


## Function defs
# Function for generating data 
def GenData(vars,x):
    amp = vars[0]
    phaseshift = vars[1]
    freq = vars[2]
    decay = vars[3]

    return amp * sin(x * freq  + phaseshift) * exp(-x*x*decay)

# residual (for optimization)
def residual(vars, x, data, eps_data):
    model = GenData(vars,x)
    return (data-model)/eps_data


## Generate data
vars = [10.0, 0.2, 3.0, 0.007]
x = np.linspace(0,1,100)
data = GenData(vars,x) + 1.*np.random.randn(np.shape(x)[0])


## fitting 
from scipy.optimize import leastsq
out = leastsq(residual, vars, args=(x, data, eps_data))
outvars = out[0]

## Compare
dataFit = GenData(outvars,x)


plt.plot(x,data,label="data")
plt.plot(x,dataFit,label="Fit")
plt.legend()
```
