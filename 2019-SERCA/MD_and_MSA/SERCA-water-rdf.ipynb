```python
def plotfigure(xdata,ydata,color="black",label='None',title='None',**font):
    matplotlib.rcParams['figure.figsize']=(2.5,1.5)
    matplotlib.rcParams['savefig.dpi'] = 300
    plt.plot(xdata,ydata,"o-",color=color,label=label,markersize=2,markeredgewidth=0.1)
    plt.xlabel("Distance ($\AA$)")
    plt.xticks(**font)
    plt.yticks(**font)
    plt.ylim([0,1])
    plt.xlim([0,12])
    plt.title(title,**font)
    plt.ylabel("Wat-Oxy density(1/$\AA^{3}$)")
```

```python
# obtain PMF from water radius distribution
# suppose the reference water density is bulk density 0.017
def getPMF(rdf,reference=0.017):
    energy=[]
    #temperature is 298 K
    RT = 0.593 
    return -RT*np.log(rdf/reference)
```

```python
# convert expt Kd to chemical potential via mu = RTlnKd, note the error propagation is
# for f=aln(bA), sigma_f = abs(a*sigma_A/A)

def from_Kd_mu(expkd):
    value=[]
    # the relation between affinity is -RTlnK = deltaG, and dossociation constant Kd = 1/K
    RT = 0.593 #kcal/mol
    mu = RT*np.log(expkd[0])
    err=RT*(expkd[1]/expkd[0]) # err
    value.append(mu)
    value.append(err)
    return value
```

```python
exp_wt_cai_mu = from_Kd_mu(exp_wt_cai)
exp_wt_caii_mu = from_Kd_mu(exp_wt_caii)
exp_wt_mg_mu = from_Kd_mu(exp_wt_mg)
exp_wt_k_mu = from_Kd_mu(exp_k)

exp_n796a_cai_mu = from_Kd_mu(exp_n796a_cai)
exp_e309q_cai_mu = from_Kd_mu(exp_e309q_cai)
exp_e309q_caii_mu = from_Kd_mu(exp_e309q_caii)
```
