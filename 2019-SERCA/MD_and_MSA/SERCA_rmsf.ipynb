```python
total_case=['wt_apo_rmsf','wt_ca_rmsf','wt_mg_rmsf','wt_k_rmsf','n796a_ca_rmsf','n796a_mg_rmsf','e309q_r1_ca_rmsf','e309q_r2_ca_rmsf']

# now use bar to illustrate the rmsf of key coordinating residues in figure
#residue numbers are: 58,309,800,796,771
selectID=771



def select_data(resid=selectID):
    error=[]
    data=[]
    data.append(wt_apo_rmsf[wt_apo_rmsf[:,0] == resid][0][1])
    data.append(wt_ca_rmsf[wt_ca_rmsf[:,0] == resid][0][1])
    data.append(n796a_ca_rmsf[n796a_ca_rmsf[:,0] == resid][0][1])
    data.append(e309q_r1_ca_rmsf[e309q_r1_ca_rmsf[:,0] == resid][0][1])
    #data.append(e309q_r2_ca_rmsf[e309q_r2_ca_rmsf[:,0] == resid][0][1])
    
    data.append(wt_mg_rmsf[wt_mg_rmsf[:,0] == resid][0][1])
    data.append(n796a_mg_rmsf[n796a_mg_rmsf[:,0] == resid][0][1])
    
    data.append(wt_k_rmsf[wt_k_rmsf[:,0] == resid][0][1])
    

   
    error.append(wt_apo_rmsf[wt_apo_rmsf[:,0] == resid][0][2])
    error.append(wt_ca_rmsf[wt_ca_rmsf[:,0] == resid][0][2])
    error.append(n796a_ca_rmsf[n796a_ca_rmsf[:,0] == resid][0][2])
    error.append(e309q_r1_ca_rmsf[e309q_r1_ca_rmsf[:,0] == resid][0][2])
    #error.append(e309q_r2_ca_rmsf[e309q_r2_ca_rmsf[:,0] == resid][0][2])
  
  
 
    error.append(wt_mg_rmsf[wt_mg_rmsf[:,0] == resid][0][2])
 
    error.append(n796a_mg_rmsf[n796a_mg_rmsf[:,0] == resid][0][2])
    error.append(wt_k_rmsf[wt_k_rmsf[:,0] == resid][0][2])

    return data,error
        
#plt.bar(ind,data,width,align='center',yerr=error,error_kw=dict(ecolor='m',lw=2,capsize=6,capthick=2))
    
```

```python
rcParams['figure.figsize']=3,2.75
rcParams['figure.dpi'] = 200
d309,e309 = select_data(resid=309)
width=0.48
xpostion=np.array([0,2,3,4,6,7,9])
ind=xpostion/2+width # now merge E309Q_r1 and E309Q_r2, only use E309Q_r1 as E309Q
print (d309)

def draw_rmsf_bar(rmsfdata,errordata,title=None,legend=False):
    barlist=plt.bar(ind,rmsfdata,width,yerr=errordata,align='center',edgecolor="black",error_kw=dict(ecolor='m',lw=2,capsize=6,capthick=2))
    colorlist=['black','red','salmon','orangered','blue','dodgerblue','green']
    for i in np.arange(len(colorlist)):
        barlist[i].set_color(colorlist[i])
    
#plt.xlabel("Simulation Time (ns)", fontsize=14)
    plt.ylabel("RMSF ($\AA$)", fontsize=14)
    plt.yticks(fontsize=13)
    plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
    plt.ylim([0,3])
    plt.title(title,fontsize=14)
    if legend:
        plt.legend([barlist[0],barlist[1],barlist[2],barlist[3],barlist[4],\
                   barlist[5],barlist[6]],["WT Apo","WT Ca","N796A Ca","E309Q Ca","WT Mg","N796A Mg","WT K"],loc=(1,0))
        #plt.legend()
```
