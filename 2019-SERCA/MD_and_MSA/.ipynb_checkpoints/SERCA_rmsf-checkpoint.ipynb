```python
import numpy as np
%matplotlib inline
import matplotlib.pylab as plt
from pylab import rcParams
from matplotlib import rc
rcParams['figure.figsize']=5,3.5
rcParams['savefig.dpi'] = 400
rc('text',usetex=True)
```

```python
case=["Apo","Ca","Mg","K","N796A"]
triplica=['run1','run2','run3']
datahome="/home/binsun/SERCA/newfigs/rmsdrmsf/"

apo_rmsd_file = datahome + case[0] + "/firmsd"
apo_rmsf_file = datahome + case[0] + "/firmsf"

print (apo_rmsf_file)
ca_rmsd_file = datahome + case[1] + "/firmsd"
ca_rmsf_file = datahome + case[1] + "/firmsf"
mg_rmsd_file = datahome + case[2] + "/firmsd"
mg_rmsf_file = datahome + case[2] + "/firmsf"
k_rmsd_file = datahome + case[3] + "/firmsd"
k_rmsf_file = datahome + case[3] + "/firmsf"

apo_rmsd = np.loadtxt(apo_rmsd_file,comments="#")
apo_rmsf = np.loadtxt(apo_rmsf_file)
        
ca_rmsd = np.loadtxt(ca_rmsd_file,comments="#")
ca_rmsf = np.loadtxt(ca_rmsf_file)

mg_rmsd = np.loadtxt(mg_rmsd_file,comments="#")
mg_rmsf = np.loadtxt(mg_rmsf_file)

k_rmsd = np.loadtxt(k_rmsd_file,comments="#")
k_rmsf = np.loadtxt(k_rmsf_file)
    
    
#data1=np.loadtxt(datafile, usecols=[0,1,2])
#cdomainrg=data[0:8,0]
#cdomaink=[0.362093, 1.02193, 1.11897, 0.498518, 0.543602, 0.291319, 0.313026, 0.35354]
#cdomainerr=[0.193402, 0.546869, 0.801458, 0.281867, 0.264447, 0.105799, 0.102451, 0.118596]

#print rg
```

```python
plt.errorbar(apo_rmsd[:,0],apo_rmsd[:,1],yerr=apo_rmsd[:,2],fmt='k-',label="Apo",linewidth=0.5,capsize=0.5)
plt.errorbar(ca_rmsd[:,0],ca_rmsd[:,1],yerr=ca_rmsd[:,2],fmt="r-",label="Ca",linewidth=0.5,capsize=0.5)
plt.errorbar(mg_rmsd[:,0],mg_rmsd[:,1],yerr=mg_rmsd[:,2],fmt="b-",label="Mg",linewidth=0.5,capsize=0.5)
plt.errorbar(k_rmsd[:,0],k_rmsd[:,1],yerr=k_rmsd[:,2],fmt="g-",label="K",linewidth=0.5,capsize=0.5)
#plt.errorbar(cdomainrg,cdomaink,yerr=cdomainerr,fmt="bo-",label="C-domain")
#plt.title("RMSD $AA$",fontsize=10)
plt.xlabel("Simulation Time (ns)", fontsize=10)
plt.ylabel("RMSD ($\AA$)", fontsize=10)

#plt.text(10.0,2.1,r'$\alpha-helical$ pCaN(native)',fontsize=8,color="red")
plt.legend(fontsize=8,)
plt.xlim([0,100])

#plt.text(10,r'$\alpha-helical$ pCaN(native)',fontsize=15,color="red")
plt.xticks(fontsize=8,weight="bold")
plt.yticks(fontsize=8,weight="bold")
```

```python
#f,(ax,ax2,ax3) = plt.subplots(1,3,sharey=True, facecolor='w')
#ax0 = f.add_subplot(111)
fig = plt.figure()
ax0 = fig.add_subplot(111)    # The big subplot to lable common axis
ax = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)
# plot the same data on both axes

ax.errorbar(apo_rmsf[:,0],apo_rmsf[:,1],yerr=apo_rmsf[:,2],fmt='k.',label="Apo",linewidth=0.5,capsize=0.3,markersize=0.3)
ax.errorbar(ca_rmsf[:,0],ca_rmsf[:,1],yerr=ca_rmsf[:,2],fmt="r.",label="Ca",linewidth=0.5,capsize=0.3,markersize=0.3)
ax.errorbar(mg_rmsf[:,0],mg_rmsf[:,1],yerr=mg_rmsf[:,2],fmt="b.",label="Mg",linewidth=0.5,capsize=0.3,markersize=0.3)
ax.errorbar(k_rmsf[:,0],k_rmsf[:,1],yerr=k_rmsf[:,2],fmt="g.",label="K",linewidth=0.5,capsize=0.3,markersize=0.3)

ax2.errorbar(apo_rmsf[:,0],apo_rmsf[:,1],yerr=apo_rmsf[:,2],fmt='k.',label="Apo",linewidth=0.5,capsize=0.3,markersize=0.3)
ax2.errorbar(ca_rmsf[:,0],ca_rmsf[:,1],yerr=ca_rmsf[:,2],fmt="r.",label="Ca",linewidth=0.5,capsize=0.3,markersize=0.3)
ax2.errorbar(mg_rmsf[:,0],mg_rmsf[:,1],yerr=mg_rmsf[:,2],fmt="b.",label="Mg",linewidth=0.5,capsize=0.3,markersize=0.3)
ax2.errorbar(k_rmsf[:,0],k_rmsf[:,1],yerr=k_rmsf[:,2],fmt="g.",label="K",linewidth=0.5,capsize=0.3,markersize=0.3)

ax3.errorbar(apo_rmsf[:,0],apo_rmsf[:,1],yerr=apo_rmsf[:,2],fmt='k.',label="Apo",linewidth=0.5,capsize=0.3,markersize=0.3)
ax3.errorbar(ca_rmsf[:,0],ca_rmsf[:,1],yerr=ca_rmsf[:,2],fmt="r.",label="Ca",linewidth=0.5,capsize=0.3,markersize=0.3)
ax3.errorbar(mg_rmsf[:,0],mg_rmsf[:,1],yerr=mg_rmsf[:,2],fmt="b.",label="Mg",linewidth=0.5,capsize=0.3,markersize=0.3)
ax3.errorbar(k_rmsf[:,0],k_rmsf[:,1],yerr=k_rmsf[:,2],fmt="g.",label="K",linewidth=0.5,capsize=0.3,markersize=0.3)




ax.set_xlim(47,118)
ax2.set_xlim(250,330)
ax3.set_xlim(750,990)
# hide the spines between ax and ax2
ax.spines['right'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax0.spines['top'].set_color('none')
ax0.spines['bottom'].set_color('none')
ax0.spines['left'].set_color('none')
ax0.spines['right'].set_color('none')
ax0.tick_params(labelcolor='w', top='off', bottom='off', left='off', right='off')

#ax2.spines['left'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax3.spines['left'].set_visible(False)

ax2.tick_params(axis='y',          # changes apply to the x-axis
                which='both',      # both major and minor ticks are affected
                left='off',      # ticks along the bottom edge are off
                right='off'
               )# ticks along the top edge are off
                #labelleft='off')
ax2.tick_params(labelright="off")
#ax.yaxis.tick_left()
ax2.tick_params(labelleft="off")
#ax2.yaxis.tick_left(False)

ax3.yaxis.tick_right()
#ax3.tick_params(labelright="off")

d = .015 # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((1-d,1+d), (-d,+d), **kwargs)
ax.plot((1-d,1+d),(1-d,1+d), **kwargs)

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d,+d), (1-d,1+d), **kwargs)
ax2.plot((-d,+d), (-d,+d), **kwargs)

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((1-d,1+d), (1-d,1+d), **kwargs)
ax2.plot((1-d,1+d), (-d,+d), **kwargs)

kwargs.update(transform=ax3.transAxes)  # switch to the bottom axes
ax3.plot((-d,+d), (1-d,1+d), **kwargs)
ax3.plot((-d,+d), (-d,+d), **kwargs)
ax3.legend()
#plt.errorbar(cdomainrg,cdomaink,yerr=cdomainerr,fmt="bo-",label="C-domain")
#plt.title("RMSD $AA$",fontsize=10)
ax0.set_xlabel("Residue ID", fontsize=10)
ax0.set_ylabel("RMSF ($\AA$)", fontsize=10)

#plt.text(10.0,2.1,r'$\alpha-helical$ pCaN(native)',fontsize=8,color="red")
#plt.legend(fontsize=8,)

#plt.text(10,r'$\alpha-helical$ pCaN(native)',fontsize=15,color="red")
#plt.xticks(fontsize=8,weight="bold")
#plt.yticks(fontsize=8,weight="bold")
```
