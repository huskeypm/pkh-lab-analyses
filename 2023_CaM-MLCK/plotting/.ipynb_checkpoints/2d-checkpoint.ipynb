```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import rcParams
from scipy.stats import ttest_ind
```

```python
systems=['mcam','cam_k30e','cam_g40d','cam_m36i','cam_t34k','cam_dN','scam1','scam4']
labels=['mCaM','K30E','G40D','M36I','T34K','$\Delta$NCaM','sCaM-1','sCaM-4']
comps=['2']
nRep=50
cols=['dodgerblue','seagreen','tab:green','limegreen','chartreuse','palegreen','crimson','lightcoral']
path='/home/xfang2/backup/faust_backup/mlck/smMLCK/md/2_mlck_cam'
```

```python
ncolors = 256
color_array = plt.get_cmap('turbo')(range(ncolors))
# change alpha values
color_array[:,-1] = np.linspace(1.0,0.0,ncolors)
color_array[0,-1] = 1
# create a colormap object
map_object = LinearSegmentedColormap.from_list(name='rainbow_alpha',colors=color_array)

ncolors = 256
color_array = plt.get_cmap('turbo')(range(ncolors))
# change alpha values
color_array[:,-1] = np.linspace(0.0,1.0,ncolors)
color_array[0,-1] = 0.0
# create a colormap object
map_object_inverted = LinearSegmentedColormap.from_list(name='rainbow_alpha',colors=color_array)
```

```python
# bottstrapping
comp=2
nRep=50
threshs=[30,5]

probs={}
for i,sys in enumerate(systems):
    probs[sys]=[]    
    for j in range(nRep):
        data=np.genfromtxt(path+'/analyses/shortMD/2d/cmDist_dfg/cm_dfg_prob_{}_{}_{}.gnu'.format(sys,comp,j+1),skip_header=7,skip_footer=3)
        zs=data[:,2]
        tot=np.sum(zs)
        n=0
        for val in data:
            x,y,z=val
            if x>=threshs[0] and y<=threshs[1]:
                n+=z
        frac=n/tot
        probs[sys].append(frac)
        
nIter=500
nSample=20
avgs=[]
stds=[]
bootData={}
for i,sys in enumerate(systems):
    data=np.array(probs[sys])
    nData=len(data)
    means=[]
    for j in range(nIter):
        idxs=np.random.choice(nData,nSample)
        draws=data[idxs]
        mean=np.mean(draws)
        means.append(mean)
    avg=np.mean(means)
    std=np.std(means)
    avgs.append(avg)
    stds.append(std)
    bootData[sys]=means
```

```python
# figs
rcParams['figure.figsize']=10,4
rcParams['font.size']=10
rcParams['figure.dpi']=300

fig=plt.figure()
for i,sys in enumerate(systems):
    ax=fig.add_subplot(2,4,i+1)
    ax.set_xlim([25, 50])
    ax.set_ylim([3, 15.5])
    ax.set_title(labels[i])
    ax.set_xlabel('CaM-MLCK distance ($\AA$)')
    ax.set_ylabel('DFG-HRD ($\AA$)')
    data=np.genfromtxt(path+'/analyses/shortMD/2d/cmDist_dfg/cm_dfg_{}_{}.gnu'.format(sys,comp),skip_header=7,skip_footer=3)
    X,Y,drawdata=data[:,0],data[:,1],data[:,2]
    z=drawdata.copy()
    z=np.append(z,z[-1])
    z=z.reshape((100,100),order='F')
    y=Y[:100]
    x=[]
    for j,val in enumerate(X):
        if j%100==0:
            x.append(val)
    ax.contour(x,y,z,cmap=map_object,vmax=np.max(drawdata),levels=5,zorder=0)  
plt.tight_layout()
plt.gcf().savefig('2d_cm_dfg.png')


rcParams['figure.figsize']=5,4
rcParams['font.size']=10
rcParams['figure.dpi']=300
fig=plt.figure()
for i,sys in enumerate(['mcam','cam_k30e','scam1','scam4']):
    ax=fig.add_subplot(2,2,i+1)
    ax.set_xlim([25, 50])
    ax.set_ylim([3, 15.5])
    ax.set_title(['mCaM','K30E','sCaM-1','sCaM-4'][i])
    ax.set_xlabel('CaM-MLCK distance ($\AA$)')
    ax.set_ylabel('DFG-HRD ($\AA$)')
    data=np.genfromtxt(path+'/analyses/shortMD/2d/cmDist_dfg/cm_dfg_{}_{}.gnu'.format(sys,comp),skip_header=7,skip_footer=3)
    X,Y,drawdata=data[:,0],data[:,1],data[:,2]
    z=drawdata.copy()
    z=np.append(z,z[-1])
    z=z.reshape((100,100),order='F')
    y=Y[:100]
    x=[]
    for j,val in enumerate(X):
        if j%100==0:
            x.append(val)
    ax.contour(x,y,z,cmap=map_object,vmax=np.max(drawdata),levels=5,zorder=0)  
plt.tight_layout()
plt.gcf().savefig('2d_cm_dfg_reduced.png')


rcParams['figure.figsize']=4,3
rcParams['font.size']=10
rcParams['figure.dpi']=300
fig=plt.figure()
ax=fig.add_subplot(111)
xs=range(len(labels))
ax.set_xticks(xs)
ax.set_xticklabels(labels,rotation=90)
ax.set_ylabel('P$_{threshold}$')
ax.set_ylim([0,0.22])
ax.bar(xs,avgs,yerr=stds,capsize=3,color=cols)
plt.tight_layout()
plt.gcf().savefig('2d_cm_dfg_bootstrap.png')
```
