```python
import numpy as np
from matplotlib import rcParams
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
```

```python
systems=['mcam','cam_k30e','cam_g40d','cam_m36i','cam_t34k','cam_dN','scam1','scam4']
labels=['mCaM', 'K30E', 'G40D', 'M36I','T34K','$\Delta$NCaM','sCaM-1', 'sCaM-4']
comp='2'
nRep=50
path='/home/xfang2/backup/faust_backup/mlck/smMLCK/md/2_mlck_cam/analyses/shortMD/dist'
start=0
end=1251
nFrame=1250
frameRate=0.2
ts=np.linspace(0,nFrame*frameRate,nFrame)
tsAll=np.linspace(0,nFrame*frameRate*nRep,nFrame*nRep)
cols=['dodgerblue','seagreen','tab:green','limegreen','chartreuse','palegreen','crimson','lightcoral']
```

```python
# CaM-MLCK distance
rcParams['figure.figsize']=8,3
rcParams['font.size']=10
thresh=30

# all reps concatenated
fig=plt.figure()    
for i,sys in enumerate(systems):
    ax=fig.add_subplot(2,4,i+1)
    ax.set_title('{} complex {}'.format(sys,comp))
    ax.set_xlabel('distance ($\AA$)')
    ax.set_ylabel('PDF')
    ax.set_ylim([0,0.4])
    ax.set_xlim([25,40])        
    dat=np.genfromtxt(path+'/cam_mlck/dist_{}_{}.dat'.format(sys,comp),skip_header=1)[:,1]
    sns.kdeplot(dat,color=cols[i])                
    kde_lines = ax.get_lines()[-1]
    kde_x, kde_y = kde_lines.get_data()
    mask = kde_x > thresh
    area=np.sum(kde_y[mask])
    filled_x, filled_y = kde_x[mask], kde_y[mask]
    # Shade the partial region 
    ax.fill_between(filled_x, y1=filled_y,color=cols[i],alpha=0.5)        
    ax.axvline(thresh,color=cols[i],linestyle='--',alpha=0.5)
plt.tight_layout()

# individual reps
probs={}
fig=plt.figure()
for i,sys in enumerate(systems):
    probs[sys]=[]
    ax=fig.add_subplot(2,4,i+1)
    ax.set_title('{} complex {}'.format(sys,comp))
    ax.set_xlabel('distance ($\AA$)')
    ax.set_ylabel('PDF')
    ax.set_ylim([0,0.4])
    ax.set_xlim([25,40]) 
    for j in range(nRep):
        dat=np.genfromtxt(path+'/cam_mlck/dist_{}_{}_{}.dat'.format(sys,comp,j+1),skip_header=1)[:,1]
        sns.kdeplot(dat,color=cols[i],alpha=0.3)
        kde_lines = ax.get_lines()[-1]
        kde_x, kde_y = kde_lines.get_data()
        mask = kde_x > thresh
        tot=np.sum(kde_y)
        prob=np.sum(kde_y[mask])
        frac=prob/tot
        probs[sys].append(frac)     
plt.tight_layout()

#bootstrapping
nIter=500
nSample=20

avgs=[]
stds=[]
for i,sys in enumerate(systems):
    data=np.array(probs[sys])
    nData=len(data)
    means=[]
    for j in range(nIter):
        idxs=np.random.choice(nData,nSample)
        draws=data[idxs]
        mean=np.mean(draws)
        means.append(mean)
    avg=np.mean(means)
    std=np.std(means)
    avgs.append(avg)
    stds.append(std)
    print(sys,avg,std)

fig=plt.figure()
ax=fig.add_subplot(111)
xs=range(len(labels))
ax.set_xticks(xs)
ax.set_xticklabels(labels,rotation=90)
ax.set_ylabel('area')
ax.bar(xs,avgs,yerr=stds,capsize=3,color=cols)

# paper fig 
rcParams['figure.figsize']=8,3
rcParams['font.size']=10
rcParams['figure.dpi']=300

fig=plt.figure()    
for i,sys in enumerate(systems):
    ax=fig.add_subplot(2,4,i+1)
    ax.set_title(labels[i])
    ax.set_xlabel('distance ($\AA$)')
    ax.set_ylabel('PDF')
    ax.set_ylim([0,0.4])
    ax.set_xlim([25,40])  
    dat=np.genfromtxt(path+'/cam_mlck/dist_{}_{}.dat'.format(sys,comp),skip_header=1)[:,1]
    sns.kdeplot(dat,color=cols[i])                
    kde_lines = ax.get_lines()[-1]
    kde_x, kde_y = kde_lines.get_data()
    mask = kde_x > thresh
    area=np.sum(kde_y[mask])
    filled_x, filled_y = kde_x[mask], kde_y[mask]
    # Shade the partial region 
    ax.fill_between(filled_x, y1=filled_y,color=cols[i],alpha=0.5)        
    ax.axvline(thresh,color=cols[i],linestyle='--',alpha=0.5)
plt.tight_layout()
plt.gcf().savefig('cmDist.png')

rcParams['figure.figsize']=4,3
rcParams['font.size']=10
rcParams['figure.dpi']=300
fig=plt.figure()
ax=fig.add_subplot(111)
xs=range(len(labels))
ax.set_xticks(xs)
ax.set_xticklabels(labels,rotation=90)
ax.set_ylabel('P($\gamma$>$\gamma_{thresh}$)')
ax.bar(xs,avgs,yerr=stds,capsize=3,color=cols)
plt.tight_layout()
plt.gcf().savefig('cmDist_bootstrap.png')
```
