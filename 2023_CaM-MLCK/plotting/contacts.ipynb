```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt
from matplotlib.ticker import FuncFormatter
import numpy as np
import matplotlib.patches as patches
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import rcParams
```

```python
systems=['mcam','cam_k30e','scam4','cam_g40d','cam_m36i','cam_t34k','cam_dN','scam1',] 
labels=['mCaM','K30E','sCaM-4','G40D','M36I','T34K','$\Delta$NCaM','sCaM-1'] 
comps=['2']
nRep=50
```

```python
# get colormap
ncolors = 256
color_array = plt.get_cmap('jet')(range(ncolors))
# change alpha values
color_array[:,-1] = np.linspace(0.5,1.0,ncolors)
color_array[0,-1] = 0
# create a colormap object
map_object = LinearSegmentedColormap.from_list(name='rainbow_alpha',colors=color_array)
```

```python
# positive values
# get colormap
ncolors = 256
color_array = plt.get_cmap('Greens')(range(ncolors))
# change alpha values
color_array[:,-1] = np.linspace(0.5,1.0,ncolors)
color_array[0,-1] = 0.0
# create a colormap object
pos_map_object = LinearSegmentedColormap.from_list(name='b',colors=color_array)

# negative values
# get colormap
ncolors = 256
color_array = plt.get_cmap('Reds')(range(ncolors))
# change alpha values
color_array[:,-1] = np.linspace(0.5,1.0,ncolors)
color_array[0,-1] = 0.0
# create a colormap object
neg_map_object = LinearSegmentedColormap.from_list(name='r',colors=color_array)
```

```python
# PCA subregions difference map
rcParams['figure.figsize']=12,12
rcParams['font.size']=15
rcParams['figure.dpi']=300
nRegions=5

for j in range(3,nRegions):
    fig=plt.figure()
    for i,sys in enumerate(systems):
        ax=fig.add_subplot(4,2,i+1)
        ax.set_title('{} area 1 versus area {}'.format(labels[i],j-1))
        ax.set_xlim([-3,144])
        ax.set_xticks([1,31,61,91,121])
        ax.set_xticklabels(['T5','L35','L65','Y95','D125'],rotation=90)
        ax.set_yticks([1460,1560,1660,1760])
        ax.set_yticklabels(['V1460','R1560','P1660','S1760'])
        #ax.set_yticks(np.arange(150,460,50))
        plt.xlabel('CaM')
        plt.ylabel('MLCK')

        # region1 data
        data=np.loadtxt('contact/pca_regions/pp_{}_2_region3'.format(sys),comments="#")
        data[:,2]=data[:,2]/np.max(data[:,2]) # normalization
        if sys == 'scam4':
            startid=144
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,444))
        elif sys == 'cam_dN':
            startid=139
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,439))
        else:
            startid=143
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,443))
        ref_data=drawdata
        
        # other regions data
        data=np.loadtxt('contact/pca_regions/pp_{}_2_region{}'.format(sys,j+1),comments="#")
        data[:,2]=data[:,2]/np.max(data[:,2]) # normalization
        
        if sys == 'scam4':
            startid=144
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,444))
            X,Y = np.meshgrid(np.arange(drawdata.shape[1]),np.arange(drawdata.shape[0]))  
            Y+=1460
        elif sys == 'cam_dN':
            startid=139
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,439))
            X,Y = np.meshgrid(np.arange(drawdata.shape[1]),np.arange(drawdata.shape[0]))  
            Y+=1460
        else:
            startid=143
            a=np.where(data[:,0] >= startid) 
            newdata= data[a[0][0]:a[0][-1]+1,:]
            drawdata=newdata[:,2].reshape((301,443))
            X,Y = np.meshgrid(np.arange(drawdata.shape[1]),np.arange(drawdata.shape[0]))  
            Y+=1460
        
        # difference
        diffData=drawdata-ref_data
        posData=diffData.copy()
        posData[diffData<=0]=0
        negData=diffData.copy()
        negData[diffData>=0]=0
        negData=abs(negData)

        plt.scatter(X,Y,c=posData,cmap=pos_map_object,s=20,marker="o",vmax=0.05) 
        plt.scatter(X,Y,c=negData,cmap=neg_map_object,s=20,marker="o",vmax=0.05) 
        plt.tight_layout()
        
    plt.savefig('contactDiff_region{}_all.png'.format(j-1))
```

```python
# get color bar
rcParams['figure.figsize']=1.5, 4
rcParams['font.size']=15

fig=plt.figure()
ax=fig.add_subplot(111)
norm=matplotlib.colors.Normalize(vmin=0, vmax=0.05)
matplotlib.colorbar.ColorbarBase(ax, cmap=pos_map_object,norm=norm)
plt.tight_layout()
plt.gcf().savefig('contactMap_colorbar_pos.png')

fig=plt.figure()
ax=fig.add_subplot(111)
norm=matplotlib.colors.Normalize(vmin=0, vmax=0.05)
matplotlib.colorbar.ColorbarBase(ax, cmap=neg_map_object,norm=norm)
plt.tight_layout()
plt.gcf().savefig('contactMap_colorbar_neg.png')
```
