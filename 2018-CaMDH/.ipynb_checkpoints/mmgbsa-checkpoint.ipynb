```python
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize']= 8,4
rcParams['figure.dpi'] = 400

import matplotlib.pylab as plt

params = {'mathtext.default': 'regular' }          
rcParams.update(params)
rcParams["font.family"] = "sans-serif"
from matplotlib.ticker import FuncFormatter
import numpy as np
```

```python
rcParams['figure.figsize']= 10,4
rcParams['figure.dpi'] = 400

colors1 = ['k','r','g']#,'m','orange','teal']

ind = np.arange(len(data[:,0]))
width=0.3
#barlist = plt.bar(ind,data[:,1],0.35,align='center',yerr=std,error_kw=dict(ecolor='m',lw=2,capsize=6,capthick=2))
barlist = plt.bar(ind,data[:,1],width,align='center')
barlist1 = plt.bar(ind+width+0.1,a454e_data[:,1],width,align='center')
#barlist2 = plt.bar(ind+2*width+0.16,g40d_data[:,1],width,align='center')

plt.plot
#plt.plot([-1,6],[10,20],"r--") 

for i,j in enumerate(barlist):
    if i <= 25:
       j.set_color('grey')
    else:
        j.set_color('k')

        
for i,j in enumerate(barlist1):
    if i <= 25:
       j.set_color('violet')
    else:
        j.set_color('fuchsia')
        
#for i,j in enumerate(barlist2):
#if i <= 25:
#       j.set_color('salmon')
#    else:
#        j.set_color('red')
#colors2 = ['grey','salmon','lightgreen']#

#for i in range(len(colors2)):
#    barlist2[i].set_color(colors2[i])
#plt.plot(1.0,53.0,'*k',ms=12)
#plt.plot(3.0,17.0,'*k',ms=12)



def residue(x, pos):
    'the two arguments take in the pos and returen residue number '
    return '%d' % (x+245+170)

formatter = FuncFormatter(residue)

#plt.xticks(ind[1:-1:3]+0.2,data[:,0][1:-1:3] + 245,fontsize=5,rotation=40)
plt.ylabel("RMSF ($\AA$)",fontsize=14)
plt.xlabel("Residue ID",fontsize=14)
#plt.legend((barlist[1],barlist[-1],("Linker","DH"))
ax = plt.gca()

ax.xaxis.set_major_formatter(formatter)
plt.xticks(fontsize=10)
plt.ylim(0,20)
#plt.xlim(-0.5,5.5)
plt.legend((barlist[1],barlist[-1],barlist1[1],barlist1[-1]),("WT-Linker","WT-DH",\
                                                                                      "A454E-Linker","A454E-DH",\
                                                                                      ),fontsize=12)
#plt.legend((barlist1[1],barlist1[-1]),("K30E-Linker","K30E-DH"),fontsize=8)
#plt.legend((barlist2[1],barlist2[-1]),("G40D-Linker","G40D-DH"),fontsize=8)
plt.savefig('aa.png')
```
