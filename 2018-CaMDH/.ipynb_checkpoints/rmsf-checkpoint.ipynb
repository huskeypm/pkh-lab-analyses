```python
%matplotlib inline
import matplotlib
import matplotlib.pylab as plt
matplotlib.rcParams['figure.figsize'] = (3,2.75)
matplotlib.rcParams['savefig.dpi'] = 300
import numpy as np
```

```python
data=[]
prename="/net/share/bsu233/CaM/Con_bding_kinetes/microsecondMD/1.5M/"
postname="/newtraj/rg/RG.dat"

name =["pCaN","lpCaN","lpcCaN"]
print name
for i,j in enumerate(name):
    print i, j
    filename = prename + j + postname
    print filename
    data.append(np.loadtxt(filename))

    
newdata=[]
newprename="/net/share/bsu233/CaM/Con_bding_kinetes/microsecondMD/0.15M/"
newpostname="/newtraj/rg/RG.dat"

newname =["pCaN","lpCaN","lpcCaN"]
print name
for i,j in enumerate(newname):
    print i, j
    filename = newprename + j + newpostname
    print filename
    newdata.append(np.loadtxt(filename))
```

```python
avg=[]
std=[]
for i,j in enumerate(data):
      avg.append(np.mean(data[i],axis=0)[1])
      std.append(np.std(data[i],axis=0)[1])
print avg,std
    
newavg=[]
newstd=[]
for i,j in enumerate(data):
      newavg.append(np.mean(newdata[i],axis=0)[1])
      newstd.append(np.std(newdata[i],axis=0)[1])
print newavg,newstd
```

```python
colors1 = ['k','r','g']#,'m','orange','teal']

ind = np.arange(len(name))
print ind
barlist = plt.bar(ind,avg,0.35,align='center',yerr=std,error_kw=dict(ecolor='m',lw=2,capsize=6,capthick=2))
barlist2 = plt.bar(ind+0.4,newavg,0.35,align='center',yerr=newstd,error_kw=dict(ecolor='m',lw=2,capsize=6,capthick=2))
plt.plot
#plt.plot([-1,6],[10,20],"r--") 

for i in range(len(colors1)):
    barlist[i].set_color(colors1[i])

colors2 = ['grey','salmon','lightgreen']#

for i in range(len(colors2)):
    barlist2[i].set_color(colors2[i])
#plt.plot(1.0,53.0,'*k',ms=12)
#plt.plot(3.0,17.0,'*k',ms=12)

plt.xticks(ind+0.2,name,weight='bold',fontsize=10,rotation=10)
plt.ylabel("Radius of Gyration ($\AA$)",weight="bold",fontsize=10)
#plt.ylim(0,60)
#plt.xlim(-0.5,5.5)

#plt.tight_layout()

#outFile = rootOutput+"Exp_Sarcolemmal_Ca_Leak.png"

#plt.gcf().savefig(outFile,dpi=300)
```
